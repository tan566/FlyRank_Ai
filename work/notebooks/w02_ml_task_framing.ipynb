## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

We are framing "Content Decline Prediction" as a **Classification** task. The goal is to predict whether a specific page will experience a significant traffic decline in the near future. We choose classification because the primary editor decision is binary: either allocate resources to refresh this page, or leave it alone.


## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target is a binary label representing decline (e.g., `is_declining`). This label comes from an **observed outcome** in the data: comparing actual search impressions in the most recent 30-day window against the previous 30-day window. It is measured reality, not a heuristic rule defined by a human.


## 3. Success metric

*One metric you can defend. What number means 'good'?*

We will use **Precision** (or Precision at K, assuming editors only have bandwidth to fix K pages a month). Precision is defensible here because the cost of a false positive—wasting hours of a subject-matter expert's time refreshing a page that didn't need it—is high. We want to be highly confident that when the model flags a page for a refresh, it is genuinely declining.


## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple fixed rule like `IF age > 1 year AND traffic is dropping THEN refresh` misses critical nuance. A page might be old but experiencing a normal seasonal dip, or it might have low search volume but high intent/conversion. ML can simultaneously weigh multiple overlapping factors (age, competition, user engagement like scroll rate, and AI traffic percentage) to find non-linear patterns that a handcrafted if-statement would miss.
